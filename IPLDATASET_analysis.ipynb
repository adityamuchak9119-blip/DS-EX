```python
def run_scored(batsman,bowler):
  mask1p=delivery['batsman']==batsman
  mask2=delivery['bowler']==bowler
  new_delivery=delivery[mask1p & mask2]
  return new_delivery.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
print(run_scored('DA Warner','TA Boult'))
```
